````python
import pandas as pd

data5 = {
    "Patient_ID":  [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008, 1009, 1010],
    "Name":        ["Ravi", "Priya", "Arjun", "Sneha", "Kumar", 
                    "Divya", "Kiran", "Meena", "Suresh", "Anita"],
    "Age":         [45, 32, None, 60, 28, 52, None, 38, 47, 55],
    "Department":  ["Cardiology", "Orthopedic", "Cardiology", "Neurology", 
                    "Orthopedic", "Cardiology", None, "Neurology", 
                    "Orthopedic", None],
    "Bill_Amount": [25000, 15000, 30000, None, 12000, 
                    28000, 22000, None, 18000, 35000],
    "Days_Admitted": [5, 3, 7, 10, 2, 6, 4, 8, 3, 9],
    "Recovered":   ["Yes", "Yes", "No", "Yes", "Yes", 
                    "No", "Yes", "No", "Yes", None]
}


##  Tasks
# ```
# SECTION A  Data Exploration
# 1. Create the DataFrame and print it
df5 = pd.DataFrame(data5)
print(df5)
print("_"*50)
# 2. Check shape  how many patients and columns
# df5.shape           #Checked

# # 3. Run .info() and .describe()
# df5.info()          #Checked
# df5.describe()      #Checked

# 4. Find how many missing values are in each column
print("Before Cleaning Data:-\n",df5.isnull().sum())
print("_"*50)

# ------------------------------------------------------------------------
# SECTION B  Data Cleaning
# 5. Fill missing Age with median age
filling_age = df5["Age"].median()
df5["Age"] = df5["Age"].fillna(filling_age)

# 6. Fill missing Bill_Amount with mean bill amount (round to 2 decimals)
filling_bill = df5["Bill_Amount"].mean()
df5["Bill_Amount"] = df5["Bill_Amount"].fillna(filling_bill).round(2)

# 7. Fill missing Department with mode
df5["Department"] = df5["Department"].fillna(df5["Department"].mode()[0])

# 8. Fill missing Recovered with mode
df5["Recovered"] = df5["Recovered"].fillna(df5["Recovered"].mode()[0])
df5

# 9. Verify zero missing values remain
checking_missing_val = df5.isnull().sum()
print("Before Cleaning Data:-\n",checking_missing_val)
print("_"*50)

# --------------------------------------------------------------------------
# SECTION C  Selecting and Filtering
# 10. Use iloc to select rows 3 to 7 and first 4 columns
select_rows_3_7 = df5.iloc[3:8 , :4]
print(select_rows_3_7)
print("_"*50)

# 11. Use loc to get Name, Department, Bill_Amount 
#     of patients admitted more than 5 days
name_dep_bill_pat = df5.loc[df5["Days_Admitted"] > 5, ["Name" , "Department", "Bill_Amount"]]
print(name_dep_bill_pat)
print("_"*50)

# 12. Filter patients who are in Cardiology department
dep_cardio_pat = df5.loc[df5["Department"] == "Cardiology"]
print(dep_cardio_pat)
print("_"*50)

# 13. Filter patients whose Bill_Amount is above 20000 
#     AND who have Recovered = "Yes"
patient_bill = df5.loc[(df5["Bill_Amount"] > 20000) & (df5["Recovered"] == "Yes")]
print(patient_bill)
print("_"*50)

# --------------------------------------------------------------------------
# SECTION D  Analysis with groupby
# 14. Find average Bill_Amount per Department
avg_bill_amount = df5.groupby("Department")["Bill_Amount"].mean()
print(avg_bill_amount)
print("_"*50)

# 15. Find total Days_Admitted per Department
total_days_dep = df5.groupby("Department")["Days_Admitted"].sum()
print(total_days_dep)
print("_"*50)

# 16. Find count of patients per Department
pat_count_dep = df5.groupby("Department")["Patient_ID"].count()
print(pat_count_dep)
print("_"*50)

# 17. Find max Bill_Amount per Department
max_bill_dep = df5.groupby("Department")["Bill_Amount"].max()
print(max_bill_dep)
print("_"*50)

# 18. Get mean, max, min of Bill_Amount 
#     per Department in ONE line using .agg()
agg_dep = df5.groupby("Department")["Bill_Amount"].agg(["mean", "max","min"])
print(agg_dep)
print("_"*50)
# --------------------------------------------------------------------------
# SECTION E  Think Like a Data Scientist
# 19. Add a new column "Bill_Per_Day" = Bill_Amount / Days_Admitted
#     (this tells cost per day for each patient)
df5["Bill_Per_Day"] = df5["Bill_Amount"] / df5["Days_Admitted"]
df5
print("_"*50)

# 20. Find which Department has the highest average Bill_Per_Day
#     Hint: groupby + mean on the new column
dep_high_billperday = df5.groupby("Department")["Bill_Per_Day"].mean().idxmax()
print(dep_high_billperday)
````

# TASK-2 BIG-REAL-WORLD PROBLEM SOLVING__PANDAS (SCHOOL DATA)

````python
import pandas as pd
import numpy as np

data6 = {
    "Student_ID":  [201, 202, 203, 204, 205, 206, 207, 208, 209, 210, 211, 212],
    "Name":        ["Aarav", "Bhavya", "Charan", "Divya", "Eshan", "Farida",
                    "Ganesh", "Hema", "Ishan", "Jaya", "Karthik", "Lavanya"],
    "Gender":      ["Male", "Female", "Male", "Female", "Male", "Female",
                    "Male", "Female", "Male", "Female", None, "Female"],
    "Math":        [85, 92, 78, None, 65, 88, 72, 95, None, 80, 70, 90],
    "Science":     [78, 88, None, 82, 70, 92, 68, 90, 75, None, 65, 85],
    "English":     [90, 85, 80, 88, None, 78, 85, 92, 70, 88, 75, None],
    "Section":     ["A", "B", "A", "B", "A", "B", "A", "B", "A", "B", None, "A"],
    "Attendance":  [95, 88, 72, 91, 65, 98, 80, 100, None, 85, 78, 92]
}


##  Tasks
# SECTION A  Data Exploration
# 1. Create the DataFrame and print it
df6 = pd.DataFrame(data6)
print(df6)
print("_"*50)

# 2. Check shape  how many students and columns
df6.shape

# 3. Run .info() and .describe()
# df6.info()
# df6.describe()

# 4. Find how many missing values in each column
print("Before Data Cleaning:-",df6.isnull().sum())
print("_"*50)
# --------------------------------------------------------------------------
# SECTION B  Data Cleaning
# 5. Fill missing Math with median
fill_math = df6["Math"].median()
df6["Math"] = df6["Math"].fillna(fill_math)

# 6. Fill missing Science with median
fill_science = df6["Science"].median()
df6["Science"] = df6["Science"].fillna(fill_science)

# 7. Fill missing English with median
fill_english = df6["English"].median()
df6["English"] = df6["English"].fillna(fill_english)

# 8. Fill missing Attendance with mean (round to 2 decimals)
fill_attendance = df6["Attendance"].mean().round(2)
df6["Attendance"]= df6["Attendance"].fillna(fill_attendance)

# 9. Fill missing Gender with mode
df6["Gender"] = df6["Gender"].fillna(df6["Gender"].mode()[0])

# 10. Fill missing Section with mode
df6["Section"] = df6["Section"].fillna(df6["Section"].mode()[0])

# 11. Verify zero missing values remain
print("After Data cleaning:-",df6.isnull().sum())
print("_"*50)
# --------------------------------------------------------------------------
# SECTION C  Feature Engineering (Creating New Columns)
# 12. Add "Total_Marks" column = Math + Science + English
df6["Total_Marks"] = df6["Math"]+df6["Science"]+df6["English"] 


# 13. Add "Average_Marks" column = Total_Marks / 3 (round to 2)
df6["Average_Marks"] = df6["Total_Marks"] / 3

# 14. Add "Grade" column using these rules:
#     Average >= 90  "A+"
#     Average >= 80  "A"
#     Average >= 70  "B"
#     Average >= 60  "C"
#     Below 60      "Fail"
#     Hint: nested np.where()  you already know this!
grade_condition = [(df6["Average_Marks"] >=90),
                   (df6["Average_Marks"] >=80),
                   (df6["Average_Marks"] >=70),
                   (df6["Average_Marks"] >=60),
                   ]
grade_choices  = ["A+" , "A", "B","C"]
df6["Grade"]= np.select(grade_condition , grade_choices ,default="Fail")    # Important operation for this kind of methods 

# 15. Add "Status" column:
#     Attendance >= 75  "Regular"
#     Attendance < 75   "Irregular"
df6["Status"] = np.where(df6["Attendance"]>=75 , "Regular" , "Irregular")
df6

# --------------------------------------------------------------------------
# SECTION D  Selecting and Filtering
# 16. Use iloc to select rows 4 to 9 and columns 1 to 5
selecting_rows = df6.iloc[4:10, 1:5]
print(selecting_rows)
print("_"*50)

# 17. Use loc to get Name, Section, Grade, Average_Marks
#     of students who scored Average_Marks above 85
stud_above_stud = df6.loc[df6["Average_Marks"] > 85 , ["Name" , "Section", "Grade","Average_Marks"]]
print(stud_above_stud)
print("_"*50)

# 18. Filter all Female students
fem_stud = df6.loc[df6["Gender"] == "Female"]
print(fem_stud)
print("_"*50)

# 19. Filter students who are Irregular in attendance
irregular_atten = df6.loc[df6["Status"] == "Irregular"]
print(irregular_atten)
print("_"*50)

# 20. Filter students from Section A 
#     whose Average_Marks is above 80
filter_secA = df6.loc[(df6["Section"] == "A") & (df6["Average_Marks"] > 80) ]
print(filter_secA)
print("_"*50)
# 21. Filter students who got Grade "A+" OR "A"
#     Hint: use .isin() function  
#     df[df["column"].isin(["value1","value2"])]
filter_stud_A = df6[df6["Grade"].isin(["A+" , "A"])]
print(filter_stud_A)
print("_"*50)
# --------------------------------------------------------------------------
# SECTION E  Groupby Analysis
# 22. Find average Math, Science, English per Section
#     in ONE line using groupby + .agg("mean") 
#     on multiple columns
find_avg_sub = df6.groupby("Section")[["Math" , "Science" ,"English"]].agg(["mean"]).round(1)
print(find_avg_sub)
print("_"*50)

# 23. Find count of students per Grade
find_count_stud = df6.groupby("Grade")["Student_ID"].count()
print(find_count_stud)
print("_"*50)

# 24. Find average Attendance per Gender
find_avg_attend = df6.groupby("Gender")["Attendance"].mean()
print(find_avg_attend)
print("_"*50)

# 25. Find highest Average_Marks per Section
find_high_marks = df6.groupby("Section")["Average_Marks"].max()
print(find_high_marks)
print("_"*50)

# --------------------------------------------------------------------------
# SECTION F  Think Like a Data Scientist
# 26. Find the name of the student with 
#     the highest Total_Marks
#     Hint: idxmax() + iloc or loc
find_high_total = df6.loc[df6["Total_Marks"].idxmax()]
print(find_high_total)
print("_"*50)

# 27. Find how many students passed (Grade != "Fail")
find_pass_stud = np.sum(df6["Grade"] != "Fail")
print(find_pass_stud)
print("_"*50)

# 28. Find what percentage of students are Irregular
#     Formula: (count of irregular / total students) * 100
count_irreg = np.sum(df6["Status"] == "Irregular")
find_perc_stud = (count_irreg / len(df6)) * 100
# print(count_irreg)
print(find_perc_stud.round(2))
print("_"*50)

# 29. Find the top 3 students by Average_Marks
#     Hint: sort_values() + head(3)
#     df.sort_values("column", ascending=False).head(3)
find_top_3 = df6.sort_values("Average_Marks" , ascending=False).head(3)
print(find_top_3)
print("_"*50)

# 30. Find which Section has better average performance
#     (higher average of Average_Marks)
avg_per = df6.groupby("Section")["Average_Marks"].mean()
high_avg = avg_per.idxmax()
print(f"The {high_avg} has the best average performance")
````
