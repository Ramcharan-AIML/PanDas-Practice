```python
# A company has 7 employees:

# Name:       ["Arun", "Bhanu", "Chitra", "Deepak",
#              "Esha", "Farhan", "Geetha"]
# Age:        [24, 35, 28, 42, 30, 26, 38]
# Salary:     [30000, 85000, 52000, 95000, 61000, 40000, 78000]
# Department: ["HR", "Tech", "Finance", "Tech",
#              "HR", "Finance", "Tech"]
# Experience: [1, 8, 4, 15, 6, 2, 10]

# Tasks:
# 1. Create a DataFrame with this data
# 2. Print first 4 rows using .head()
# 3. Check shape, info, and describe
# 4. Select only the Name and Salary columns
# 5. Filter employees whose salary is above 60000
# 6. Filter employees who are in the Tech department
# 7. Add a new column "Bonus" = 10% of salary for everyone
# 8. Find the average salary using the DataFrame column
```

```python
import pandas as pd
# import numpy as np

data = {
    "Name" :      ["Arun", "Bhanu", "Chitra", "Deepak", "Esha", "Farhan", "Geetha"],
    "Age" :       [24, 35, 28, 42, 30, 26, 38],
    "Salary":     [30000, 85000, 52000, 95000, 61000, 40000, 78000],
    "Department": ["HR", "Tech", "Finance", "Tech", "HR", "Finance", "Tech"],
    "Experience": [1, 8, 4, 15, 6, 2, 10]
}


df = pd.DataFrame(data)

# Tasks:
# 1. Create a DataFrame with this data
print(df)
print("_"*45)
# 2. Print first 4 rows using .head()
first_4_rows = df.head(4)
print(first_4_rows)

# 3. Check shape, info, and describe
# print(df.shape)
# print("_"*45)
# df.describe()
# print("_"*45)
# df.info()

# 4. Select only the Name and Salary columns
name_col = df[["Name","Salary"]]
print(name_col)
print("_"*45)

# 5. Filter employees whose salary is above 60000
filter_sal = df[df["Salary"] > 60000]
print(filter_sal)
print("_"*45)

# 6. Filter employees who are in the Tech department
filter_tech = df[df["Department"] == "Tech"]
print(filter_tech)
print("_"*45)

# 7. Add a new column "Bonus" = 10% of salary for everyone
df["Bonus Salary"] = df["Salary"] * 1.10
print(df)

# 8. Find the average salary using the DataFrame column
avg_salary = df["Salary"].mean()
print(avg_salary)
```

```python
import numpy as np

data2 = {
    "Name":       ["Ravi", "Priya", "Arjun", "Sneha", "Kumar", "Divya"],
    "Age":        [28, 32, None, 29, 35, 27],
    "Salary":     [45000, None, 50000, 80000, None, 60000],
    "Department": ["HR", "Tech", None, "Tech", "HR", "Finance"]
}
df2 = pd.DataFrame(data2)
print(df2)

print(df2.isnull())
print(df2.isnull().sum())

df_dropped = df2.dropna()
print(df_dropped)


# fillna()
filling_age = df2["Age"].mean()
df2["Age"] = df2["Age"].fillna(filling_age)

filling_salary = df2["Salary"].median()
df2["Salary"] = df2["Salary"].fillna(filling_salary)


df2["Department"] = df2["Department"].fillna(df2["Department"].mode()[0])

print(df2)
print(df2.isnull().sum())


# Grouby operation
grouping = df2.groupby("Department")["Salary"].mean()
print(grouping)

print(df2.groupby("Department")["Salary"].sum())
print(df2.groupby("Department")["Salary"].count())
print(df2.groupby("Department")["Salary"].max())
print("_"*50)
print(df2.groupby("Department")["Salary"].agg(["mean" , "max" , "count" ,"sum"]))
print(df2.groupby("Department")[["Salary", "Age"]].mean())


```
